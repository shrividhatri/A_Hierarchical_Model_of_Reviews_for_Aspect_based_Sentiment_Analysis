```python
import csv
```

```python
arr=[]
array=[]
array2=[]
with open('Data1.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        line_count += 1
        x=0
        for a in row:
            arr.append(a)
        for i in row:
            if i=="category":
                array.append(x)
            if i=="polarity":
                array2.append(x)
            x+=1
            
print(array)
print(array2)
```

```python
text=[]
aspect=[]
pol=[]
for num in range(414):
    text.append([arr[0+115*num],arr[13+115*num],arr[20+115*num],arr[27+115*num],arr[64+115*num],arr[71+115*num],arr[78+115*num]])
    pol.append([[arr[3+115*num],arr[10+115*num]],[arr[16+115*num]],[arr[23+115*num]],[arr[30+115*num],arr[36+115*num],arr[41+115*num],arr[46+115*num],arr[51+115*num],arr[56+115*num],arr[61+115*num]],[arr[67+115*num]],[arr[74+115*num]],[arr[81+115*num],arr[87+115*num],arr[92+115*num],arr[97+115*num],arr[102+115*num],arr[107+115*num],arr[112+115*num]]])
    aspect.append([[arr[2+115*num],arr[9+115*num]],[arr[15+115*num]],[arr[22+115*num]],[arr[29+115*num],arr[35+115*num],arr[40+115*num],arr[45+115*num],arr[50+115*num],arr[55+115*num],arr[60+115*num]],[arr[66+115*num]],[arr[73+115*num]],[arr[80+115*num],arr[86+115*num],arr[91+115*num],arr[96+115*num],arr[101+115*num],arr[106+115*num],arr[111+115*num]]])
```

```python
data =[]
num=[]
abc=0
for i in text:
    if abc==0:
        abc+=1
        continue;
    rev=[]
    print(i)
    for j in i:
        words=[]
        if j=='':
            for k in range(35):
                words.append("EOF")
        else:
            words=j.split()
            num.append(len(words))
            for k in range(len(words),35):
                words.append("EOF")
        rev.append(words)
    for ab in range(len(rev),7):
        words=[]
        for k in range(35):
            words.append("EOF")
        rev.append(words)
    data.append(rev)
    abc+=1
```

```python
import numpy as np
data = np.array(data)
```

```python
category=[]
for i in aspect:
    asp=[]
    if i[0][0]=='category':
        continue;
    for j in i:
        sent=[]
        for k in j:
            if k !='':
                sent.append(k)
        asp.append(sent)
    category.append(asp)
```

```python
polar=[]
for i in pol:
    ar1=[]
    if i[0][0]=='polarity':
        continue;
    for j in i:
        ar2=[]
        for k in j:
            if k !='':
                ar2.append(k)
        ar1.append(ar2)
    polar.append(ar1)
```

```python
aspects = np.array(category)
```

```python
polarity = np.array(polar)
```
